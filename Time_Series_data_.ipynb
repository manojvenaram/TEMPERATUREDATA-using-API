```python
import pandas as pd
import random
data=pd.read_csv("/content/cities_r2.csv")
data1=data["name_of_city"]
# random_names = data1.sample(n=100, random_state=1)
L=list(data1)
```

```python
import requests

# Your API key
api_key = '5fe926db79707ac9e25439e62b64fc71'
d={}
# Get the city name from the user
# city = input("Enter the city name: ")
for i in L:
# API endpoint
  url = f'http://api.openweathermap.org/data/2.5/weather?q={i}&appid={api_key}'

  # Make a GET request
  response = requests.get(url)

  # Check if the request was successful
  if response.status_code == 200:
      # Parse JSON response
      data = response.json()
      # Extract relevant information
      temperature = data['main']['temp']
      weather_description = data['weather'][0]['description']
      d[i]=temperature
      print(f'Temperature: {temperature}K')
      print(f'Weather Description: {weather_description}')
  else:
      print(f'Error: {response.status_code}')
```
